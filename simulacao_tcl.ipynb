# Definindo a Funcao

```python
#bibliotecas utilizadas
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.style.use('bmh')

"""
definindo a funcao responsavel por gerar arrays aleatorios contendo
os resultados das jogadas de um dado nao viciado e gerar um histograma
com os resultados

"""

def teocentral(tam_amostral, n_amostras, bins=6, lower=1, upper=7):
    lista = []
    for i in range(n_amostras):
        x = np.random.randint(lower, upper, size= tam_amostral).mean()
        lista.append(x)
    fig, ax = plt.subplots(figsize=(5,3))
    ax.hist(lista, bins=bins,color='#FEEA00',edgecolor='black', linewidth=.5)
    ax.set_ylabel('Frequencia')
    ax.set_xlim(lower-1, upper)
```

```python
teocentral(1,10000)
```
