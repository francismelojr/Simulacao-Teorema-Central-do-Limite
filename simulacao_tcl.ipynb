# Simulacao do Teorema Central do Limite

* Importando bibliotecas e definindo a Funcao

```python
#bibliotecas utilizadas
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

"""
definindo a funcao responsavel por gerar arrays aleatorios contendo
os resultados das jogadas de um dado nao viciado e gerar um histograma
com a densidade das frequÃªncias
"""

def teocentral(tam_amostral, n_amostras, bins=10, lower=1, upper=7):
    lista = []
    for i in range(n_amostras):
        x = np.random.randint(lower, upper, size= tam_amostral).mean()
        lista.append(x)
    plt.figure(figsize= (5,3))
    ax = sns.histplot(lista, bins=bins, color='#FEEA00',edgecolor='black', linewidth=0.5, stat='density')
    ax = sns.kdeplot(lista, bw_adjust=4, color= 'blue')
    ax.set_ylabel('Densidade')
    ax.set_xlim(0, 7)
```

* Simulando o experimento para um tamanho amostral 10.000 e numero de amostras variavel

```python
#n=1
teocentral(1,10000, bins=6)
```

```python
#n=2
teocentral(2,10000)
```

```python
#n=5
teocentral(5,10000)
```

```python
#n=10
teocentral(10,10000)
```

```python
#n=50
teocentral(50,10000)
```

```python
#n=100
teocentral(100,10000)
```
