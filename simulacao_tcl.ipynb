# Simulacao do Teorema Central do Limite

```python
#packages
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

"""
defining a function responsible for generating random arrays containing
the results of rolls of an unbiased die and generate a histogram
with the frequency density
"""

def central_limit(sample_size, n_samples, bins=10, lower=1, upper=7):
    lista = []
    for i in range(n_samples):
        x = np.random.randint(lower, upper, size= sample_size).mean()
        lista.append(x)
    plt.figure(figsize= (8,6))
    ax = sns.histplot(lista, bins=bins, color='#FEEA00',edgecolor='black', linewidth=0.5, stat='density')
    ax = sns.kdeplot(lista, bw_adjust=4, color= 'blue')
    ax.set_ylabel('Density')
    ax.set_xlim(0, 7)
```

```python
#n=1
central_limit(1,10000, bins=6)
```

```python
#n=2
central_limit(2,10000)
```

```python
#n=5
central_limit(5,10000)
```

```python
#n=10
central_limit(10,10000)
```

```python
#n=50
central_limit(50,10000)
```

```python
#n=100
central_limit(100,10000)
```
